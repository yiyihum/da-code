```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('../data.csv')
```

```python
df.shape
```

```python
# the last row is null, lets remove it as well
df = df.dropna()
df.tail()
```

```python
df['Profit'] = df['Revenue'] - df['Cost']
df['Profit Margins'] = round(df['Profit']/df['Revenue'] * 100, 2)
df['Unit Profit']=df['Unit Price']-df['Unit Cost']
df['Unit Profit Margin']= round(df['Unit Profit']/df['Unit Price'] * 100 , 2)
df
```

```python
# 计算各州在所有年份中的总利润并排序
state_profit = df.groupby(['State'])['Profit'].sum().sort_values(ascending=False)

# 保存各州总利润到CSV文件
state_profit.to_csv('state_profit.csv', header=['Total Profit'])

print("Calculated data and state profit CSV files have been generated successfully.")
```
