```python
rating_columns = ['Overall Rating', 'Seat Comfort', 'Staff Service', 'Food & Beverages']
for col in rating_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Remove rows with NaN values in any of the rating columns
df = df.dropna(subset=rating_columns)

# Group by Airline and calculate mean ratings for overall satisfaction and specific aspects
airline_ratings = df.groupby('Airline')[rating_columns].mean().reset_index()


airline_ratings
```

```python

rating_columns = ['Overall Rating', 'Seat Comfort', 'Staff Service', 'Food & Beverages']
for col in rating_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna(subset=rating_columns)

correlation_matrix = df[rating_columns].corr()

print("Correlation Matrix:")
print(correlation_matrix)

correlation_matrix.to_csv('result.csv')
```
